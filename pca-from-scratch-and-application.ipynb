Steps to perform PCA(Principal Component Analysis):

1. Standardize the data

2. Compute the covariance matrix

3. Compute eigen values and eigne vector

4. Sort the eigen values

5. Claculte variance explained and cumulative variance

6. select number of principle components

7. Interpret the data

 **Color vlaues**

0 - black

max(255) - white

It will fuctuate a lot. So we need to standize the data inorder to make a smooth flow between balck and white color

```python
from sklearn.preprocessing import StandardScaler

# Standardization: scaling down the data in a way where mean=0, std deviation = 1
```

```python
x_std = StandardScaler().fit_transform(x)
x_std
```

```python
x_std.shape
```

```python
# Transpose: rows to columns
# We take transpose because np.cov(covariance) expects rows as features
x1 = x_std.T
x1
```

```python
x1.shape
```

```python
# covariance: tell us how features vary w.r.t to each other
cov_mat = np.cov(x1)
cov_mat
```

```python
cov_mat.shape
```

```python
eig_val, eig_vec = np.linalg.eig(cov_mat)
# eigen vector: direction theought which we get the max. variance
# eighen values: explain the variance
```

```python
eig_val
```

```python
eig_vec
```

```python
len(eig_val)
```

```python
len(eig_vec)
```

```python
total = sum(eig_val)
total
```

```python
# variance explainex by each component in percentage
# list of percentage variance explained by each pricipal component
var_exp=[]
for i in eig_val:
  percentage=(i/total)*100
  var_exp.append(percentage)
```

```python
var_exp
```

```python
# cumulative variance
cum_sum_var = np.cumsum(var_exp)
```

```python
cum_sum_var
```

```python
# var_exp -> helps you see the contribution of each principal component
# cum_sum_var -> helps you decide how many PCs to keep

# example: if first 2 pcs explain almost 90% variance, yo don't need the rest
```

```python
plt.figure(figsize=(10,5))
plt.bar(range(len(var_exp)), var_exp, label='Individual explained variance', color='g')
plt.step(range(len(cum_sum_var)), cum_sum_var, label = 'Cumulative explained variance')
plt.ylabel('Explained variance')
plt.xlabel('Principal components')
plt.legend()
plt.show()
```

# Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA

pca = PCA(n_components = 21)  # from the graph decided by us
pca_x = pca.fit_transform(x_std)
```

```python
pca_x.shape   # 64 columns are reduced to 21
```

```python
x_train,x_test,y_train,y_test = train_test_split(pca_x,y,train_size=0.8,random_state=23)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state = 23)
dt.fit(x_train, y_train)
```

```python
predict = dt.predict(x_test)
predict
```

```python
y_test
```

```python
from sklearn.metrics import *

accuracy_score(y_test, predict)
```
